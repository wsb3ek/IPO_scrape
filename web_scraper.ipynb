```python
import requests # pip install requests
import urllib.request
import time
import datetime
import random
from datetime import timedelta
import bs4 # pip install beautifulsoup4 or conda install -c conda-forge bs4
from bs4 import BeautifulSoup
import json
import pprint
import re
import pandas as pd
```

```python
# scraping the IPO calendar from Marketwatch.com
url = "https://www.marketwatch.com/tools/ipo-calendar"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
tabPanes_body = soup.find('div','element__body j-tabPanes')
tabPanes = tabPanes_body.find_all('div', 'tab__pane')
tabPanes
```

```python
dfs = {}
current_datetime = datetime.datetime.now().strftime('%m-%d-%y %H_%M_%S')
current_datetime
```

```python
xlsxwriter = pd.ExcelWriter('IPO Calendar ({0}).xlsx'.format(current_datetime))
```

```python
for tabPane in tabPanes:
    headers = tabPane.find_all('header')
    for indx, header in enumerate(headers):
        tab_name = header.h2.div.contents[0]
        dfs[tab_name] = pd.read_html(str(tabPane.find_all('table', {'class': 'table'})[indx]))[0]
        dfs[tab_name].to_excel(xlsxwriter, sheet_name=tab_name, index=False)
```

```python
xlsxwriter.save()
```
